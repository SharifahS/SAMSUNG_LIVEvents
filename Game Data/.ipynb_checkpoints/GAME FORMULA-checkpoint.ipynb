```python
import pandas as pd
import numpy as np 
```

```python
data = pd.read_csv('Desktop\monday-results.csv')
#pd.set_option('display.max_rows', 100)
```

```python
data = data.fillna(method="ffill")
data
```

```python
#First aggregation calculating PERCENT & # OF QUESTIONS ANSWERED
dgroup4 = data.groupby([ "NAME", 'User ID',
                        "Email", "Room #", 'Rank By Room', 
                        'User Score'])['True/False'].agg(['mean', 'count']).reset_index()
```

```python
# Second aggregation for START TIME & END TIME
dgroup45 = data.groupby([ "NAME", 'User ID',
                        "Email", "Room #", 'Rank By Room', 
                        'User Score'])['Date & Time'].agg(['min', 'max']).reset_index()
```

```python
# creating a new column to add the TOTAL QUESTIONS
Questions_Per_Room = []
    
for row in dgroup45['Room #']:
    if row == 7.0 :    Questions_Per_Room.append('10')
    elif row == 26.0:   Questions_Per_Room.append('3')
    elif row == 4.0:   Questions_Per_Room.append('1')
    elif row == 8.0:   Questions_Per_Room.append('10')
    elif row == 29.0:   Questions_Per_Room.append('3')
    elif row == 30.0 :  Questions_Per_Room.append('1')
    elif row == 33.0:   Questions_Per_Room.append('3')
    elif row == 35.0:   Questions_Per_Room.append('3')
    elif row == 36.0:   Questions_Per_Room.append('1')
    elif row == 39.0:   Questions_Per_Room.append('3')
    elif row == 41.0 :  Questions_Per_Room.append('3')
    elif row == 42.0:   Questions_Per_Room.append('1')
    elif row == 44.0:   Questions_Per_Room.append('3')
    elif row == 46.0:   Questions_Per_Room.append('3')
    elif row == 48.0:   Questions_Per_Room.append('3')
    elif row == 50.0 :  Questions_Per_Room.append('1')
    elif row == 57.0:   Questions_Per_Room.append('1')
    elif row == 59.0:   Questions_Per_Room.append('3')
    elif row == 52.0:   Questions_Per_Room.append('3')
    elif row == 25.0:   Questions_Per_Room.append('1')    
    elif row == 32.0:   Questions_Per_Room.append('1')
    elif row == 38.0:   Questions_Per_Room.append('1') 
    elif row == 45.0:   Questions_Per_Room.append('1')    
        #elif row = 54.0:   Questions_Per_Room.append('3')    
        
    else:           Questions_Per_Room.append('3')
    
       
```

```python
#Adding QUESTION PER ROOM to dataset
dgroup4['Questions_Per_Room'] = Questions_Per_Room
```

```python
# Take created columns from second aggregation 
# ****only run once****
dgroup46 = dgroup45[['min', 'max']]


#join columns to first aggregation
dgroup4 = pd.merge(dgroup4, dgroup46, left_index=True, right_index=True)
```

```python
# Calling PERCENTAGE column & # OF QUESTIONS ANSWERED
dgroup5 = dgroup4[dgroup4.columns[6]]
dgroup6 = dgroup4[dgroup4.columns[7]]

#How many True & How many false per game per person calculations 
trueframe = dgroup5 * dgroup6
falseframe = dgroup6 - trueframe
```

```python
# Adding new columns  TOTAL CORRECT & TOTAL INCORRECT to dataset(agggregation1)
dgroup4['Total Correct']= trueframe
dgroup4['Total Incorrect'] = falseframe
```

```python
#Column rename
dgroup4 = dgroup4.rename(columns= {'mean': 'Percent',
                                  'count':'# of Questions Answered',
                                   'Questions_Per_Room': "Total Questions Per Room",
                                  'min':'Start Time',
                                  'max':"End Time"})
```

```python
#Third aggregatING ROWS Collecting all User Answers
dgroup0 = data.groupby([ "NAME", 'User ID',
                        "Email", "Room #", 'Rank By Room', 
                        'User Score'], as_index = False).agg({'User Answer': ' , '.join})
```

```python
# Get column from 3rd aggregation
# ****only run once****
UserA = dgroup0['User Answer']

#join column to dataset(aggregation 1)
DATA = pd.merge(dgroup4, UserA, left_index=True, right_index=True)
 

```

```python
# Find all negative characters

#DATA['Game Status'] = np.where(DATA['# of Questions Answered'] == DATA['Questions_Per_Room'], False, True)
```

```python
data80 = DATA['# of Questions Answered']
data90 = DATA['Total Questions Per Room']
data70 =  DATA['Total Correct']

# Numeric type function allowing you to calculate columns 
data80 = pd.to_numeric(data80) 
data90 = pd.to_numeric(data90)
data70 = pd.to_numeric(data70)

#Calculate PERCENT column and GAME STATUS column
PERCENT = data70 / data90 * 100
Game_Status = data90 - data80

#Add new Columns to Dataset
DATA['PERCENT%'] = PERCENT
DATA['Game Status'] = Game_Status
```

```python
# Fill in GAME STATUS w/completed and incomplete
#number replace represents total missed in Game 
DATA['Game Status'] = DATA['Game Status'].replace([0], 'Complete')
DATA['Game Status'] = DATA['Game Status'].replace([1], 'Incomplete')
DATA['Game Status'] = DATA['Game Status'].replace([9], 'Incomplete')
DATA['Game Status'] = DATA['Game Status'].replace([4], 'Incomplete')
DATA['Game Status'] = DATA['Game Status'].replace([2], 'Incomplete')
DATA['Game Status'] = DATA['Game Status'].replace([3], 'Incomplete')
DATA['Game Status'] = DATA['Game Status'].replace([5], 'Incomplete')
DATA['Game Status'] = DATA['Game Status'].replace([7], 'Incomplete')
DATA['Game Status'] = DATA['Game Status'].replace([6], 'Incomplete')
DATA['Game Status'] = DATA['Game Status'].replace([8], 'Incomplete')
```

```python
# sorting Room #
#DATA = DATA.sort_values('Room #', ascending=True)
```

```python
#DATA.set_index(['Start Time','End Time'])
#DATA.set_index('Room #')
DATA = DATA.sort_values(['Room #','Rank By Room'], ascending=True)
```
